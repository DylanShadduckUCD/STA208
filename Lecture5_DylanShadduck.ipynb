```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import preprocessing, model_selection, linear_model
```

```python
filename = "/Users/Dylan/Desktop/School Stuff/Master's Year/Spring 2021/STA 208/DavisSML/data/Hitters.csv"
df = pd.read_csv(filename, index_col=0).dropna()
```

```python
df.head()
```

```python
df.index.name = "Player"
dummies = pd.get_dummies(df[['League', 'Division', 'NewLeague']])

# We want to predict salary, so we will store this as our predictor variable y
y = df.Salary

# Drop unncessary columns from our data frame
X_ = df.drop(['Salary', 'League', 'Division', 'NewLeague'], axis=1).astype('float64')

# Add back our dummy columns
X = pd.concat([X_, dummies[['League_N', 'Division_W', 'NewLeague_N']]], axis=1)
```

```python
X.head()
```

```python
# Split our X and y into training and test sets
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.33, random_state=0)
```

```python
loo = model_selection.LeaveOneOut()
looiter = loo.split(X)
lasso = linear_model.LassoCV(cv=looiter, max_iter=2000, random_state=0)
```

```python
lasso.fit(X, y)
lasso.get_params()
```

```python
print("The selected lambda value is {:.2f}".format(lasso.alpha_))
```

```python
print(lasso.coef_)
```
